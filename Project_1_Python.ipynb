# Rock-Paper-Scissors Game

Welcome to the Rock-Paper-Scissors game! You can play against the computer by selecting one of the buttons below (Rock, Paper, or Scissors). The computer will randomly pick one as well, and the result will be displayed right here. Have fun!


```python
import random
import ipywidgets as widgets
from IPython.display import display, clear_output

# Raw GitHub URLs for the images in your repository
rock_image_url = "https://raw.githubusercontent.com/MAMDOUH-ALSHARARI/Test_Project_1_Python/main/images/rock.png"
paper_image_url = "https://raw.githubusercontent.com/MAMDOUH-ALSHARARI/Test_Project_1_Python/main/images/paper.png"
scissors_image_url = "https://raw.githubusercontent.com/MAMDOUH-ALSHARARI/Test_Project_1_Python/<branch>/images/scissors.png"

# Choices for the game
choices = ["rock", "paper", "scissors"]

# Create an Output widget to display the result message
output = widgets.Output()

# Function to determine the winner and return the result and color
def determine_winner(player_choice, computer_choice):
    if player_choice == computer_choice:
        return "It's a tie!", 'orange'
    elif (player_choice == "rock" and computer_choice == "scissors") or \
         (player_choice == "scissors" and computer_choice == "paper") or \
         (player_choice == "paper" and computer_choice == "rock"):
        return "You win!", 'green'
    else:
        return "Computer wins!", 'red'

# Function to play the game and display the result
def play_game(player_choice):
    computer_choice = random.choice(choices)
    result_message, color = determine_winner(player_choice, computer_choice)

    # Clear previous output and display the new result
    with output:
        clear_output()
        print(f"You chose: {player_choice.capitalize()}")
        print(f"Computer chose: {computer_choice.capitalize()}")
        display(widgets.HTML(f"<h2 style='color:{color}'>{result_message}</h2>"))

# Create buttons with images using HTML in the description
rock_button = widgets.Button(
    description=f"<img src='{rock_image_url}' width='100' height='100'>",
    layout=widgets.Layout(width='150px', height='150px'),
    tooltip='Rock'
)
paper_button = widgets.Button(
    description=f"<img src='{paper_image_url}' width='100' height='100'>",
    layout=widgets.Layout(width='150px', height='150px'),
    tooltip='Paper'
)
scissors_button = widgets.Button(
    description=f"<img src='{scissors_image_url}' width='100' height='100'>",
    layout=widgets.Layout(width='150px', height='150px'),
    tooltip='Scissors'
)

# Set up button click events
rock_button.on_click(lambda x: play_game("rock"))
paper_button.on_click(lambda x: play_game("paper"))
scissors_button.on_click(lambda x: play_game("scissors"))

# Display the buttons and the result output
buttons_box = widgets.HBox([rock_button, paper_button, scissors_button])
display(buttons_box, output)
```
