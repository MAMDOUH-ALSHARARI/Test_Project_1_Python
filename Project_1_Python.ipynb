# Rock-Paper-Scissors Game

Welcome to the Rock-Paper-Scissors game! You can play against the computer by selecting one of the buttons below (Rock, Paper, or Scissors). The computer will randomly pick one as well, and the result will be displayed right here. Have fun!


```python
import random
import ipywidgets as widgets
from IPython.display import display, clear_output

# Choices for the game
choices = ["rock", "paper", "scissors"]

# Initialize scores and rounds
player_score = 0
computer_score = 0
current_round = 1
total_rounds = 0

# Create an Output widget to display the result message
output = widgets.Output()

# Function to determine the winner and return the result and color
def determine_winner(player_choice, computer_choice):
    if player_choice == computer_choice:
        return "It's a tie!", 'orange'
    elif (player_choice == "rock" and computer_choice == "scissors") or \
         (player_choice == "scissors" and computer_choice == "paper") or \
         (player_choice == "paper" and computer_choice == "rock"):
        return "You win!", 'green'
    else:
        return "Computer wins!", 'red'

# Function to play the game and update scores
def play_game(player_choice):
    global player_score, computer_score, current_round, total_rounds

    computer_choice = random.choice(choices)
    result_message, color = determine_winner(player_choice, computer_choice)

    # Update scores based on the result
    if result_message == "You win!":
        player_score += 1
    elif result_message == "Computer wins!":
        computer_score += 1

    # Display round result
    with output:
        clear_output()
        print(f"Round {current_round}/{total_rounds}")
        print(f"You chose: {player_choice.capitalize()}")
        print(f"Computer chose: {computer_choice.capitalize()}")
        display(widgets.HTML(f"<h2 style='color:{color}'>{result_message}</h2>"))
        print(f"Score -> You: {player_score} | Computer: {computer_score}")

    # Check if the game is over
    current_round += 1
    if current_round > total_rounds:
        end_game()

# Function to end the game and display the final result
def end_game():
    with output:
        clear_output()
        if player_score > computer_score:
            display(widgets.HTML(f"<h2 style='color:green'>You won the game! Final score: You {player_score} - Computer {computer_score}</h2>"))
        elif computer_score > player_score:
            display(widgets.HTML(f"<h2 style='color:red'>Computer won the game! Final score: You {player_score} - Computer {computer_score}</h2>"))
        else:
            display(widgets.HTML(f"<h2 style='color:orange'>It's a tie! Final score: You {player_score} - Computer {computer_score}</h2>"))

# Function to start a new game with the specified number of rounds
def start_game(rounds):
    global player_score, computer_score, current_round, total_rounds
    player_score = 0
    computer_score = 0
    current_round = 1
    total_rounds = int(rounds)

    # Clear the output and instructions
    with output:
        clear_output()
        print(f"Game started! Playing {total_rounds} rounds.")

# Create buttons for user choices
rock_button = widgets.Button(description="Rock", layout=widgets.Layout(width='150px', height='50px'))
paper_button = widgets.Button(description="Paper", layout=widgets.Layout(width='150px', height='50px'))
scissors_button = widgets.Button(description="Scissors", layout=widgets.Layout(width='150px', height='50px'))

# Set up button click events
rock_button.on_click(lambda x: play_game("rock"))
paper_button.on_click(lambda x: play_game("paper"))
scissors_button.on_click(lambda x: play_game("scissors"))

# Create a widget for user to input the number of rounds
round_input = widgets.BoundedIntText(
    value=3,  # Default value
    min=1,    # Minimum number of rounds
    max=10,   # Maximum number of rounds
    step=1,
    description='Rounds:',
    layout=widgets.Layout(width='200px')
)

# Button to start the game
start_button = widgets.Button(description="Start Game", layout=widgets.Layout(width='150px'))

# Set up start button event
start_button.on_click(lambda x: start_game(round_input.value))

# Create layout for the input and start button
input_box = widgets.HBox([round_input, start_button])

# Display the input box, buttons, and output
display(input_box)
display(widgets.HBox([rock_button, paper_button, scissors_button]))
display(output)
```
